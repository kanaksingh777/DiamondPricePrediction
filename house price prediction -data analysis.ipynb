```python
#checking for missing values 
```

```python
missing_values = df.isna().mean()*100
```

```python
#finding the columns with missing values 
```

```python
missing_features = missing_values[missing_values>0]
```

```python
missing_features
```

```python
#making the list of these columns 
```

```python
missing_features = missing_features.index.tolist()
```

```python
missing_features 
```

```python
#finding the relationship between these missing feature columns with dependent variable sales price 
```

```python
for feature in missing_features:
    data= df.copy()
    data[feature]= np.where(data[feature].isnull(),1,0)
    
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

```python
#finding numerical variable 
```

```python
numerical_features = []
for feature in df.columns:
    if df[feature].dtype != 'O':
        numerical_features.append(feature)

print('number of numerical features are:',len(numerical_features))
```

```python
df[numerical_features].head()
```

```python
df[numerical_features].shape
```

```python
#there are some features that are data time variable (temporal variable)--- year feature y
```

```python
year_feature =[]
for feature in numerical_features:
    if 'Yr' in feature or 'Year' in feature:
        year_feature.append(feature)
```

```python
year_feature
```

```python
#finding discrete and continuous numerical variables
```

```python
discrete_feature = []
for feature in numerical_features:
    if len(df[feature].unique())<25 and feature not in  year_feature+ ['Id']:
     discrete_feature.append(feature)

print('no of discrete_features are:' ,len(discrete_feature))
```

```python
continuous_feature = []
for feature in numerical_features:
    if feature not in discrete_feature+year_feature +['Id']:
        continuous_feature.append(feature)

print('no of continuous feature are {}'.format(len(continuous_feature)))
```

```python
df[continuous_feature].head()
```

```python
#finding the distribution of continuous features with histogram
```

```python
for feature in continuous_feature:
    data = df.copy()
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.show()
    
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python

for feature in continuous_feature:
    data = df.copy()
    sns.distplot(data[feature], kde=True, bins=25)
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.show()
```

```python
#we need to normalize continuous feature in gaussian distribution for linear models 
#skewed data is transformed with the help of logarithmic transformation
```

```python
for feature in continuous_feature:
    data = df.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature] = np.log(data[feature])
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalesPrice')
        plt.title(feature)
        plt.show()
 
```

```python
#dealing with outliers
```

```python
for feature in continuous_feature:
    data= df.copy()
    data[feature] = np.log(data[feature])
    plt.boxplot(data[feature])
    plt.ylabel(feature)
    plt.show()
```
